```python
import folium
from folium import plugins

# Data lokasi di Semarang
lokasi_semarang = [
    {"nama": "Kalibanteng", "lat": -6.9845739281732975, "lon": 110.38351440522527},
    {"nama": "Madukoro", "lat": -6.980879592814979, "lon": 110.39905219218345},
    {"nama": "Indraprasta", "lat": -6.97857137309681, "lon": 110.41163595541559},
    {"nama": "Tugu Muda", "lat": -6.983552591745231, "lon": 110.41007168405561},
    {"nama": "Kyai Saleh", "lat": -6.986660206981591, "lon": 110.41393529540247},
    {"nama": "Bergota", "lat": -6.993292725367926, "lon": 110.4137406704881},
    {"nama": "Kaligarang", "lat": -6.995766634139341, "lon": 110.40231267560914}
]

# Hitung pusat peta (rata-rata koordinat)
avg_lat = sum([lok["lat"] for lok in lokasi_semarang]) / len(lokasi_semarang)
avg_lon = sum([lok["lon"] for lok in lokasi_semarang]) / len(lokasi_semarang)

# Buat peta dengan pusat di Semarang
peta = folium.Map(
    location=[avg_lat, avg_lon],
    zoom_start=13,
    tiles='OpenStreetMap'
)

# Tambahkan marker untuk setiap lokasi
for idx, lokasi in enumerate(lokasi_semarang, 1):
    folium.Marker(
        location=[lokasi["lat"], lokasi["lon"]],
        popup=f"<b>{lokasi['nama']}</b><br>Lat: {lokasi['lat']:.6f}<br>Lon: {lokasi['lon']:.6f}",
        tooltip=lokasi["nama"],
        icon=folium.Icon(color='red', icon='info-sign', prefix='glyphicon')
    ).add_to(peta)
    
    # Tambahkan circle marker untuk penanda lebih jelas
    folium.CircleMarker(
        location=[lokasi["lat"], lokasi["lon"]],
        radius=8,
        popup=lokasi["nama"],
        color='#3186cc',
        fill=True,
        fillColor='#3186cc'
    ).add_to(peta)

# Tambahkan marker cluster (opsional, bisa diaktifkan jika banyak titik)
# marker_cluster = plugins.MarkerCluster().add_to(peta)

# Tambahkan fullscreen button
plugins.Fullscreen(
    position='topright',
    title='Fullscreen',
    title_cancel='Exit Fullscreen',
    force_separate_button=True
).add_to(peta)

# Tambahkan measure control untuk mengukur jarak
plugins.MeasureControl(position='topleft').add_to(peta)

# Simpan peta ke file HTML
peta.save('peta_semarang.html')

print("Peta berhasil dibuat! Buka file 'peta_semarang.html' di browser Anda.")
print(f"\nTotal lokasi: {len(lokasi_semarang)}")
print("\nDaftar lokasi:")
for idx, lokasi in enumerate(lokasi_semarang, 1):
    print(f"{idx}. {lokasi['nama']}: ({lokasi['lat']:.6f}, {lokasi['lon']:.6f})")
```
